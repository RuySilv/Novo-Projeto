# Médias
---

```python
#Média(Casos Confirmados)
Medconfirmado = covid['Confirmed'].mean()
print("A média de casos confirmados é: %.2f" %Medconfirmado)

#Média(Mortes)
Medmorte = covid['Deaths'].mean()
print("A média de mortes é: %.2f" %Medmorte)

#Média(Recuperados)
Medrec = covid["Recovered"].mean()
print("A média de recuperados é: %.2f" %Medrec)

#Média(Casos Ativos)
Medact = covid["Active"].mean()
print("A média de recuperados é: %.2f" %Medact)

#Média(Morte para cada 100 casos)
Medmp100 = covid["Deaths / 100 Cases"].mean()
print("A média de Mortes a cada 100 casos é: %.2f" %Medmp100)

#Média(Recuperados para cada 100 casos)
Medrecp100 = covid["Recovered / 100 Cases"].mean()
print("A média de Recuperados a cada 100 casos é: %.2f" %Medrecp100)

#Média (Confirmados na última semana)
Medlw = covid["Confirmed last week"].mean()
print("A média de confirmados na última semana é: %.2f" %Medlw)

#Média(% de aumentos em uma semana)
Medincrease = covid["1 week % increase"].mean()
print("A média de %% de aumentos em uma semana é: %.2f" %Medincrease)
```

Os gráficos serão retirados do dataframe (covid).

```python
# Gráfico referente ao número de mortes nos países das Americas (Top6).

region1 = covid[covid["WHO Region"] == "Americas"]
aux1 = region1.sort_values(by=['Deaths'], ascending=False).head(6)
aux1
aux1.drop(columns=['New cases',"New deaths","New recovered","1 week change"])
```

```python
ax = sns.barplot(data=aux1, x="Deaths", y="Country/Region")
ax.set(xlabel="Número de mortes", ylabel="Países")
```

```python
# Gráfico referente ao número de mortes nos países da Africa (Top 6).
region2 = covid[covid["WHO Region"] == "Africa"]
aux2 = region2.sort_values(by=['Deaths'], ascending=False).head(6)
aux2
aux2.drop(columns=['New cases',"New deaths","New recovered","1 week change"])
```

```python
sns.lineplot(data=aux2, x="Country/Region", y="Active",marker="o")
```

```python
# Gráfico referente ao número de mortes nos países da Europa (Top 6).

region3 = covid[covid["WHO Region"] == "Europe"]
aux3 = region3.sort_values(by=['Deaths'], ascending=False).head(6)
aux3
aux3.drop(columns=['New cases',"New deaths","New recovered","1 week change"])
```

```python
# Gráfico de correlação
x = aux3[["Deaths / 100 Cases","Recovered / 100 Cases","Deaths / 100 Recovered"]].corr()
sns.heatmap(x, annot=True)
```

```python
# Gráfico referente ao número de casos confirmados nos países da Western Pacific (Top 6).
region4 = covid[covid["WHO Region"] == "Western Pacific"]
aux4 = region4.sort_values(by=['Deaths'], ascending=False).head(6)
aux4
aux4.drop(columns=['New cases',"New deaths","New recovered","1 week change"])
```

```python
y = aux4.groupby("Country/Region").sum()

y.plot.pie(y="Confirmed")
```
