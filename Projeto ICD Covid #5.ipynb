# Agrupamentos
---

```python
#Recuperados a cada 100 casos e Mortes a cada 100 casos.
df = covid[["Country/Region","Recovered / 100 Cases","Deaths / 100 Cases","WHO Region"]]
dfnovo = covid[["Recovered / 100 Cases","Deaths / 100 Cases"]]
#Este DataFrame(dfnovo) serÃ¡ usado no Kmeans, logo nÃ£o poderÃ¡ conter Strings(paÃ­ses).
df
```

```python
sns.scatterplot(data=df,x="Deaths / 100 Cases",y="Recovered / 100 Cases");
#Como a dispersÃ£o acontece mais prÃ³xima ao eixo X , isso significas que sÃ£o recuparadas 
# mais pessoas do que pessoas morrem, ambos a cada 100 casos.
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7)
kmeans.fit(dfnovo)
```

```python
sns.scatterplot(data=df,x="Deaths / 100 Cases",y="Recovered / 100 Cases", hue=kmeans.labels_);
```

```python
finalk = df[kmeans.labels_== 1]
finalk
```

```python
pd.value_counts(finalk["WHO Region"])
```
