**3º) Utilizando a técnica de regressão linear, é possível identificar a quantidade de mortes em relação ao número de recuperados, ambos a cada 100 casos ?**

### Regressão Linear

---

<p> A Regressão Linear é usado para análises predativas através de algoritmos supervisionados de Machine Learning. Logo, seu principal objetivo é localizar as melhores estimativas para os coeficientes, minimizando os erros na previsão de y (variável dependente) a partir de x (variável independente).

```python
from scipy.stats import linregress
```

```python
# Separando os parâmetros para a regressão linear

result = linregress(covid["Recovered / 100 Cases"], covid["Deaths / 100 Cases"])
result
```

```python
# Seria o slope do result (coeficiente angular)

a = result.slope
a
```

```python
# Seria o intercept do result (coeficiente linear)

b = result.intercept
b
```

```python
x = covid["Recovered / 100 Cases"]
y = covid["Deaths / 100 Cases"]
cont = y.count()
```

```python
# A função show_regression abaixo serve para visualizar o resultado da regressão linear.

import matplotlib.pyplot as plt

def gen_line(x_min, x_max, n_points, a, b):
    interval = x_max-x_min
    interval_value = interval/n_points
    
    x_result = []
    y_result = []

    for i in range(n_points):
        current_x = x_min+i*interval_value
        x_result.append(current_x)
        y_result.append(a*current_x + b)
        
    return x_result, y_result

def show_regression(x, y, a, b):
    N = 100

    plt.scatter(x, y)

    line_1 = gen_line(min(x), max(x), N, a, b)
    plt.plot(line_1[0], line_1[1], c= "red")

    plt.gca()
    plt.show()
```

```python
gen_line(x[0],x[15],cont,a,b)
```

```python
show_regression(x, y, a, b)
```

```python
def predict(x):
    y_predict = b + a*x
    
    return y_predict
```

```python
predict(0)
```

```python
predict(50)
```

```python
predict(100)
```

><p> Portanto, com base no algoritmo, sendo "Recovered / 100 Cases" a variável dependente e "Deaths / 100 Cases" a variável independente, nota-se que quanto maior o número de recuperados a cada 100 casos, menor é a quantidade de mortes.
